```python
import random
# used the random library to generate a random number to be guessed by the user
```

```python
#create a function called "guess"
def guess(x):
  random_no=random.randint(1,x)  # generate a random no.
  guess= int(input(f"Guess a number b/w 1 to {x}: ")) #take the guess value input from the user
  while (guess!= random_no) :                        # using while loop, guess is not equal to rnadom_no, it will check if its lower or higher

    if guess < random_no:
      print("Ohno! you guessed too low. Try again.")

    elif guess > random_no:
      print("Ohno! you guessed too high. Try again.")

    guess= int(input(f"Guess a number b/w 1 to{x}: ")) # take another guess, if the guess value is lower/higher

  print("yay! you guessed it correctly.")              # if guess= random_no, print correctly guessed.
```

```python
n=int(input("Enter upper limit : ")) # take the limit

guess(n) # call the function, with the limit enetered as the parameter.
```
