```python
import random
```

```python
def comp_guess(x,y):
  low=x
  high=y
  feedback=''
  while feedback!='c':
    if low!=high:
      guess= random.randint(low,high)
    else:
      guess=low
    feedback=input(f"Is {guess} too high(H), too low(L) or correct(C)? ").lower()
    if feedback=='h':
      high=guess-1
    elif feedback=='l':
      low= guess +1

  print(f"Yay, your guess {guess} is correct.")
```

```python
a=int(input("Enter the lower limit of the number: "))
b=int(input("Enter the higher limit of the number: "))
comp_guess(a,b)
```
